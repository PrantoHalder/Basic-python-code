```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as splt
import seaborn
```

```python
df=pd.read_csv('Weather Report.csv')#,sep=';') #importing the dataset
df #showing the data set that is uploded
```

```python
df.shape #condition of the data
```

```python
df.isnull().values.any()  #null value check
```

```python
df.isna().sum() #column wise null value condition
```

```python
df=df.drop(['SL No'],axis=1)
df=df.drop(['Station'],axis=1)
df=df.drop(['Year'],axis=1)
df
```

```python
df.corr()
splt.figure(figsize=(16,10))

splt.xticks(fontsize=14)
splt.yticks(fontsize=14)
seaborn.heatmap(df.corr(),annot=True)
splt.show()
```
